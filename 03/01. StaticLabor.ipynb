# Static Labor Supply

# Setup

```python
import numpy as np
from scipy.optimize import minimize

import matplotlib.pyplot as plt

from consav.linear_interp import interp_1d
```

## Utility and Parameters

```python
# utility
def util(h,w,N,par):
    C = income(h,w,N,par) #vi medregner budgetbetingelsen hvorfor nyttefunktionen ikke afhænger af C
    return (C**(1.0+par.eta))/(1.0+par.eta) - par.beta*(h**(1.0+par.gamma))/(1.0+par.gamma)

# income
def income(h,w,N,par):
    return (1.0-par.tau)*w*h + N

# solution
def solve(N,par):

    h = np.nan + np.ones(par.w_grid.size) #Højredelen er en vektor af et taller hvor venstredelen er en vektor af scalar.
    #h initialiseres med NaN, fordi det senere skal overskrives med optimerede værdier. 
    #Der oprettes en array h af samme størrelse som par.w_grid, hvor hver værdi er NaN. 
    #Dette bruges til at gemme de optimerede værdier for arbejdskrafttilbuddet for hver værdi af lønnen w
    for iw,w in enumerate(par.w_grid): #Her itereres over alle værdier af w i par.w_grid. Indekset iw bruges til at gemme den optimerede værdi af h i arrayet h.
        obj = lambda h: -util(h,w,N,par) #definerer optimeringsproblem
        h_init = np.array([2.0]) #Der vælges en startværdi på 2 for arbejdskrafttilbuddet h, som minimize() starter fra.

        bounds = ((0.0,200.0),) #Arbejdskrafttilbuddet h begrænses til at ligge mellem 0 og 200. 
        #Dette sikrer, at løsningen ikke bliver urealistisk lav eller høj.

        res = minimize(obj,h_init,bounds=bounds,method='SLSQP') #Metoden SLSQP (Sequential Least Squares Quadratic Programming) er valgt, da den håndterer både begrænsninger (bounds) og differentierbare funktioner godt.
        h[iw] = res.x[0] #Den optimerede værdi af h for den aktuelle w gemmes i arrayet h.
        
    return h
```

```python
# parameters 
class par: pass
par.beta = 0.1
par.eta = -2.0
par.gamma = 2.5
par.tau = 0.1
```

## Solve Static Model on Grid of Wages, $w$

```python
# find optimal labor supply as a function of wages
par.w_grid = np.linspace(1.0,20.0,100)

N = 5.0 # non-labor income. Important for strength of income effect
sol_h = solve(N,par)
```

```python
# plot solution
fig, ax = plt.subplots()
ax.plot(par.w_grid,sol_h,marker='o');
ax.set(xlabel='wages, $w$',ylabel='Hours, $h^{\star}$');
```

## Simulation
To simulate from this model, we need a level of non-labor income, $N$, and a wage level, $w$. We will primarely focus on the latter. 

The simulation here is simple, since we just need to *interpolate* the found solution for the new value of $w$ which is likely not on the grid `par.w_grid`, for which we have calculated the solution.

```python
def simulate(wi,sol_h,par): #a level of wages and a vector of optimal labor supply
    return interp_1d(par.w_grid,sol_h,wi)
```

## Elasticities
**Marshall**

```python
N = 5.0 # non-labor income. Important for strength of income effect
sol_h = solve(N,par)
# Theoretical Marshall elasticities
S = ((1-par.tau)*par.w_grid*sol_h) / ((1-par.tau)*par.w_grid*sol_h + N)
ela_M = (1+par.eta*S) / (par.gamma - par.eta*S)
```

```python
# Simulate one point on this curve. One observation is enough because no heterogeneity
w = 5.0
h_sim_base = simulate(w,sol_h,par) 
h_sim_1pct = simulate(w*1.01,sol_h,par) 

ela_M_sim = (h_sim_1pct-h_sim_base)/h_sim_base*100
```

```python
# plot
fig, ax = plt.subplots()
ax.plot(par.w_grid,ela_M,label='Theoretical');
ax.scatter(w,ela_M_sim,color='red',label='Simulated');
ax.set(xlabel='wages, $w$',ylabel='Marshall, $e_{M}$');
ax.legend();
```

**Income effect**

```python
# Theoretical
ei = par.eta*S/(par.gamma - par.eta*S) 
```

```python
# Solve and simulate model with slightly higher non-labor income
h_N = solve(N*1.01,par)
h_sim_1pct_N = simulate(w,h_N,par)

ela_I_sim = (h_sim_1pct_N-h_sim_base)/h_sim_base*100
ei_sim = w*h_sim_base*(1-par.tau)* ela_I_sim / N
```

```python
# plot
fig, ax = plt.subplots()
ax.plot(par.w_grid,ei,label='Theoretical');
ax.scatter(w,ei_sim,color='red',label='Simulated');
ax.set(xlabel='wages, $w$',ylabel='Income effect, $ei$');
ax.legend();
```

**Hicks, compensated elasticity**

```python
# Theoretical
ela_H = 1/(par.gamma - par.eta*S)
```

```python
# Hicks: Compensate (or use Slutsky, which is much easier)
def util_diff(comp,h_base,w,N,par):
    
    util_base = util(h_base,w,N,par)
    
    # alternative model: increase wage by 1% but compensate via non-labor income
    N_alt = N + comp
    w_alt = w*1.01
    
    h_alt_vec = solve(N_alt,par)
    h_alt = simulate(w_alt,h_alt_vec,par)
    util_alt = util(h_alt,w_alt,N_alt,par)
    
    return util_base - util_alt
```

```python
# find the compensation that equalize utility to baseline level
obj = lambda comp: 1000.0*util_diff(comp,h_sim_base,w,N,par)**2 # squared utility difference (scaled)
res = minimize(obj,0.0)
comp = res.x[0]
#obj definerer en funktion, der minimerer forskellen i nytte før og efter lønændringen.
#1000.0 * er blot en skalering for at gøre optimeringen mere numerisk stabil.
#minimize(obj, 0.0) forsøger at finde den comp, der sætter forskellen i nytte lig 0.

print(f'Compensation: {comp:2.4f}. Resulting utility difference: {util_diff(comp,h_sim_base,w,N,par):2.2g}')
```

```python
# calculate elasticity with this level of compensation
h_H = solve(N+comp,par)
h_sim_1pct_H = simulate(w*1.01,h_H,par)
ela_H_sim = (h_sim_1pct_H-h_sim_base)/h_sim_base*100
```

```python
# plot
fig, ax = plt.subplots()
ax.plot(par.w_grid,ela_H,label='Theoretical');
ax.scatter(w,ela_H_sim,color='red',label='Simulated');
ax.set(xlabel='wages, $w$',ylabel='Hicks, $e_H$');
ax.legend();
```

```python
# WRITE YOUR CODE HERE
# Number of agents
num_agents = 1000

# Generate random wages uniformly between 2.5 and 10.0
np.random.seed(2023)
wages = np.random.uniform(2.5, 10.0, num_agents)

# Compute baseline labor supply
h_baseline = np.array([solve(N, par)[np.argmin(np.abs(par.w_grid - w))] for w in wages])

# Compute labor supply at a slightly higher wage w + Δw
delta_w = 0.01  # Small increase in wages
h_new = np.array([solve(N, par)[np.argmin(np.abs(par.w_grid - (w + delta_w)))] for w in wages])

# Compute Marshall elasticity: ε^M = (dh/dw) * (w/h)
marshall_elasticities = ((h_new - h_baseline) / delta_w) * (wages / h_baseline)

# Compute the average Marshall elasticity
average_marshall_elasticity = np.nanmean(marshall_elasticities)

# Print the result
print(f"Average Marshall Elasticity: {average_marshall_elasticity:.4f}")
```
